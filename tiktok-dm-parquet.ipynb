```python
import os
import zipfile



def unzip_files_in_directory(target_dir):
    # Iterate over each file in the target directory
    for file_name in os.listdir(target_dir):
        if file_name.endswith(".zip"):
            file_path = os.path.join(target_dir, file_name)
            # Define a directory to extract files into
            extract_dir = os.path.join(target_dir, os.path.splitext(file_name)[0])
            os.makedirs(extract_dir, exist_ok=True)
            
            try:
                # Attempt to unzip the file
                with zipfile.ZipFile(file_path, 'r') as zip_ref:
                    zip_ref.extractall(extract_dir)
                    print(f"Extracted {file_path} to {extract_dir}")
            except zipfile.BadZipFile:
                print(f"Error: {file_path} is not a valid ZIP file.")
            except Exception as e:
                print(f"An unexpected error occurred while extracting {file_path}: {e}")



# Unzip all files in the specified directory
unzip_files_in_directory(target_directory)
```

```python
import os
import shutil

def move_direct_messages_text_to_root_and_cleanup(target_dir):
    for sub_dir_name in os.listdir(target_dir):
        sub_dir_path = os.path.join(target_dir, sub_dir_name)
        
        # Check if it's a directory
        if os.path.isdir(sub_dir_path):
            # Define the path to the 'Direct Messages' text file
            direct_messages_file_path = os.path.join(sub_dir_path, "Tiktok", "Direct Messages", "Direct Messages.txt")
            
            # Check if the 'Direct Messages' text file exists
            if os.path.isfile(direct_messages_file_path):
                # Define the target path in the root of the subdirectory
                target_file_path = os.path.join(sub_dir_path, "Direct Messages.txt")
                
                # Copy the 'Direct Messages' text file to the root of the subdirectory
                shutil.copy2(direct_messages_file_path, target_file_path)
                print(f"Copied 'Direct Messages.txt' to root of '{sub_dir_path}'")
                
                # After copying, delete everything else in the subdirectory except the copied file
                for item_name in os.listdir(sub_dir_path):
                    item_path = os.path.join(sub_dir_path, item_name)
                    # Skip the copied 'Direct Messages.txt' file
                    if item_name == "Direct Messages.txt":
                        continue
                    # Remove all other contents
                    if os.path.isdir(item_path):
                        shutil.rmtree(item_path)
                    elif os.path.isfile(item_path):
                        os.remove(item_path)

                print(f"Cleaned up all other files in '{sub_dir_path}'.")



# Execute the function
move_direct_messages_text_to_root_and_cleanup(target_directory)
```

```python
def load_txt_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = file.readlines()
    return data

def load_txt_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()  # Read all lines
    return lines

def extract_urls_with_timestamps(lines):
    url_data = []
    for line in lines:
        # Use regex to search for URLs
        url_pattern = r'(https?://\S+)'  # Pattern for URLs (both http and https)
        match = re.search(url_pattern, line)
        if match:
            # Extract timestamp (assuming it's the first part of the line)
            timestamp = line.split()[0] + " " + line.split()[1]
            url = match.group(1)
            url_data.append({'timestamp': timestamp, 'url': url})
    return url_data

def save_urls_to_file(url_data, output_file):
    with open(output_file, 'w', encoding='utf-8') as file:
        for entry in url_data:
            file.write(f"{entry['timestamp']} - {entry['url']}\n")


def parse_messages(messages):
    parsed_data = []
    
    for line in messages:
        # Regex to match timestamp (e.g., "2023-09-10 02:17:14 sender: message")
        match = re.match(r'(\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}) ([^:]+): (.+)', line)
        if match:
            timestamp, sender, message = match.groups()
            parsed_data.append({
                'timestamp': timestamp,
                'sender': sender,
                'message': message.strip()
            })
    
    return parsed_data

def anonymize_senders(parsed_data):
    sender_mapping = {}
    counter = 1
    
    for msg in parsed_data:
        sender = msg['sender']
        if sender not in sender_mapping:
            sender_mapping[sender] = f'User_{counter}'
            counter += 1
        msg['sender'] = sender_mapping[sender]
    
    return parsed_data, sender_mapping

# def mask_sensitive_content(parsed_data):
#     url_pattern = re.compile(r'http[s]?://\S+')
    
#     for msg in parsed_data:
#         # Replacing URLs with [REDACTED]
#         msg['message'] = re.sub(url_pattern, '[REDACTED]', msg['message'])
    
#     return parsed_data





def process_and_store_dm_data(dm_data, output_dir):
    # Parse the DM data into a DataFrame
    df = pd.DataFrame(dm_data)
    # Add 'is_message_url' column
    url_pattern = re.compile(r'https?://\S+')
    df['is_message_url'] = df['message'].apply(lambda x: 1 if url_pattern.search(x) else 0)
    
    # Add empty 'transcription' and 'alt_text' columns
    df['transcription'] = ""
    df['alt_text'] = ""
    
    # Define the output Parquet file path
    output_file_path = os.path.join(output_dir, "DMs.parquet")
    
    # Save DataFrame to Parquet file
    df.to_parquet(output_file_path, compression = 'gzip', index=False)
    print(f"Parquet file created at: {output_file_path}")



```

```python
import pandas as pd
import requests
import torch
from moviepy.editor import VideoFileClip
from transformers import BlipProcessor, BlipForConditionalGeneration
from subprocess import call
import numpy as np
import os
import re
from torchvision.ops import nms

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)

# Initialize BLIP model for image-to-text
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to(device)

def sanitize_filename(url):
    """
    Sanitizes the TikTok URL to create a safe filename.
    Removes or replaces characters that are invalid in filenames.
    """
    sanitized = re.sub(r'[^\w\s-]', '_', url)  # Replace non-alphanumeric characters with underscores
    return sanitized

def download_tiktok_video(url):
    """
    Downloads a TikTok video and returns the local file path.
    """
    try:
        output_path = sanitize_filename(url) + ".mp4"
        result = call(["yt-dlp", url, "-o", output_path])
        if result == 0 and os.path.exists(output_path):
            print(f"Video downloaded successfully: {output_path}")
            return output_path
        else:
            print(f"Video download failed for {url}")
            return None
    except Exception as e:
        print(f"Error downloading video: {e}")
        return None


def extract_frames_at_intervals(video_path, interval=30):
    """
    Extracts frames at a specified interval from the video.
    """
    clip = VideoFileClip(video_path)
    frames = []

    frame_rate = clip.fps
    frame_interval = interval / frame_rate

    for t in np.arange(0, clip.duration, frame_interval):
        frame = clip.get_frame(t)
        frames.append(frame)

    return frames

def generate_alt_text(image):
    """
    Uses BLIP to generate descriptive alt text from a single frame.
    """
    inputs = processor(image, return_tensors="pt").to(device)
    with torch.no_grad():
        output = model.generate(**inputs)
    description = processor.decode(output[0], skip_special_tokens=True)
    return description

import pandas as pd

def is_tiktok_video_url(url):
    """
    Checks if the provided URL is a TikTok video URL.
    """
    is_tiktok = "/video/" in url
    print(f"Checking URL: {url}, Is TikTok: {is_tiktok}")
    return is_tiktok


def process_parquet_with_alt_text(input_parquet, output_parquet):
    # Load the Parquet file
    df = pd.read_parquet(input_parquet)

    # Ensure 'alt_text' column exists
    if 'alt_text' not in df.columns:
        df['alt_text'] = None

    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        if row['is_message_url'] == 1:
            url = row['message']
            if is_tiktok_video_url(url):
                try:
                    # Step 1: Download the video
                    video_path = download_tiktok_video(url)

                    # Step 2: Extract frames
                    frames = extract_frames_at_intervals(video_path, interval=30)

                    # Step 3: Generate alt text for each frame
                    alt_texts = [generate_alt_text(frame) for frame in frames]
                    combined_alt_text = " ".join(alt_texts)  # Combine descriptions

                    # Step 4: Update the 'alt_text' field
                    df.at[index, 'alt_text'] = combined_alt_text
                except Exception as e:
                    print(f"Error processing URL at index {index}: {e}")
                    df.at[index, 'alt_text'] = "Error generating alt text"
            else:
                # Not a video URL, set 'alt_text' to -1
                df.at[index, 'alt_text'] = -1

    # Ensure the 'alt_text' column has consistent data types
    df['alt_text'] = df['alt_text'].astype(str).fillna("")

    # Save the modified DataFrame back to Parquet
    df.to_parquet(output_parquet, compression = 'gzip')

# Example usage
# process_parquet_with_alt_text("input_file.parquet", "output_file.parquet")


def process_all_dms_in_subdirectories(target_directory):
    for root, dirs, files in os.walk(target_directory):
        for file in files:
            if file == 'DMs.parquet':
                input_parquet_path = os.path.join(root, file)
                output_parquet_path = os.path.join(root, 'DMs.parquet')
                
                print(f"Processing {input_parquet_path}...")
                process_parquet_with_alt_text(input_parquet_path, output_parquet_path)
                print(f"Saved processed file to {output_parquet_path}")

# Example usage


```

```python
import whisper
from subprocess import call

# Initialize Whisper model
whisper_model = whisper.load_model("base")

def extract_audio(video_path, audio_path):
    """
    Extract audio from a video file using ffmpeg.
    """
    try:
        print(f"Extracting audio from video: {video_path}")
        call(["ffmpeg", "-i", video_path, "-q:a", "0", "-map", "a", audio_path, "-y"])
        if os.path.exists(audio_path):
            print(f"Audio extracted successfully: {audio_path}")
            return audio_path
        else:
            print(f"Audio extraction failed for video: {video_path}")
            return None
    except Exception as e:
        print(f"Error extracting audio from {video_path}: {e}")
        return None


def transcribe_audio(audio_path):
    """
    Transcribes audio using Whisper.
    """
    try:
        print(f"Transcribing audio file: {audio_path}")
        result = whisper_model.transcribe(audio_path)
        transcription = result.get("text", "")
        print(f"Transcription: {transcription}")
        return transcription
    except Exception as e:
        print(f"Error transcribing audio: {e}")
        return "Error generating transcription"

def is_tiktok_video_url(url):
    """
    Checks if the provided URL is a TikTok video URL.
    """
    is_tiktok = "/video/" in url
    print(f"Checking URL: {url}, Is TikTok: {is_tiktok}")
    return is_tiktok


def process_parquet_with_transcription(input_parquet, output_parquet):
    # Load the Parquet file
    df = pd.read_parquet(input_parquet)

    # Ensure 'transcription' column exists
    if 'transcription' not in df.columns:
        df['transcription'] = None

    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        if row['is_message_url'] == 1:
            url = row['message']
            print(f"Processing row {index} with URL: {url}")
            
            if is_tiktok_video_url(url):
                try:
                    # Step 1: Download the video
                    video_path = download_tiktok_video(url)
                    if not video_path:
                        df.at[index, 'transcription'] = "Error downloading video"
                        continue

                    # Step 2: Extract audio
                    audio_path = sanitize_filename(url) + ".mp3"
                    if not extract_audio(video_path, audio_path):
                        df.at[index, 'transcription'] = "Error extracting audio"
                        continue

                    # Step 3: Transcribe the audio
                    transcription = transcribe_audio(audio_path)
                    print(f"Row {index} Transcription: {transcription}")  # Debugging

                    # Step 4: Update the 'transcription' field
                    if transcription:
                        df.at[index, 'transcription'] = transcription
                    else:
                        df.at[index, 'transcription'] = "Empty transcription"

                except Exception as e:
                    print(f"Error processing row {index}: {e}")
                    df.at[index, 'transcription'] = "Error generating transcription"
            else:
                # Not a video URL
                print(f"Row {index} is not a TikTok URL.")
                df.at[index, 'transcription'] = -1

    # Ensure the 'transcription' column has consistent data types
    df['transcription'] = df['transcription'].astype(str).fillna("")

    # Save the modified DataFrame back to Parquet
    df.to_parquet(output_parquet)



# Example usage
# process_parquet_with_transcription("input_file.parquet", "output_file.parquet")

def process_all_dms_with_transcription(target_directory):
    for root, dirs, files in os.walk(target_directory):
        for file in files:
            if file == 'DMs.parquet':
                input_parquet_path = os.path.join(root, file)
                output_parquet_path = os.path.join(root, 'DMs.parquet')
                
                print(f"Processing {input_parquet_path}...")
                process_parquet_with_transcription(input_parquet_path, output_parquet_path)
                print(f"Saved processed file to {output_parquet_path}")

# Example usage

```

```python
import os
import pandas as pd
import whisper
from subprocess import call

# Initialize Whisper model
whisper_model = whisper.load_model("base")

import requests
from bs4 import BeautifulSoup

def download_tiktok_music(url):
    """
    Downloads TikTok music audio from a given music URL by scraping the audio file link.
    """
    try:
        print(f"Downloading audio from TikTok music URL: {url}")
        
        # Follow the redirect and scrape the page for the audio URL
        response = requests.get(url, allow_redirects=True)
        if response.status_code != 200:
            print(f"Failed to fetch TikTok music page: {url}")
            return None

        soup = BeautifulSoup(response.text, "html.parser")

        # Look for the audio file URL in the page source
        audio_tag = soup.find("meta", property="og:audio")  # TikTok might embed it as og:audio
        if not audio_tag or not audio_tag.get("content"):
            print("No audio file link found on the TikTok music page.")
            return None

        audio_url = audio_tag["content"]
        print(f"Found audio file URL: {audio_url}")

        # Download the audio file
        audio_file_name = sanitize_filename(url) + ".mp3"
        audio_response = requests.get(audio_url, stream=True)
        if audio_response.status_code == 200:
            with open(audio_file_name, "wb") as f:
                for chunk in audio_response.iter_content(chunk_size=8192):
                    f.write(chunk)
            print(f"Audio downloaded successfully: {audio_file_name}")
            return audio_file_name
        else:
            print(f"Failed to download audio file from {audio_url}")
            return None

    except Exception as e:
        print(f"Error downloading audio from {url}: {e}")
        return None

def sanitize_filename(url):
    """
    Creates a safe filename based on the URL.
    """
    return url.split("/")[-1].split(".")[0]

def transcribe_audio(audio_path):
    """
    Transcribes audio using Whisper.
    """
    try:
        print(f"Transcribing audio file: {audio_path}")
        result = whisper_model.transcribe(audio_path)
        transcription = result.get("text", "")
        print(f"Transcription: {transcription}")
        return transcription
    except Exception as e:
        print(f"Error transcribing audio: {e}")
        return "Error generating transcription"

def is_tiktok_music_url(url):
    """
    Checks if the provided URL is a TikTok music URL.
    """
    is_music = "/music/" in url
    print(f"Checking URL: {url}, Is TikTok Music: {is_music}")
    return is_music

def process_parquet_with_music_transcription(input_parquet, output_parquet):
    # Load the Parquet file
    df = pd.read_parquet(input_parquet)

    # Ensure 'transcription' column exists
    if 'transcription' not in df.columns:
        df['transcription'] = None

    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        if row['is_message_url'] == 1:
            url = row['message']
            print(f"Processing row {index} with URL: {url}")
            
            if is_tiktok_music_url(url):
                try:
                    # Step 1: Download the music audio
                    audio_path = download_tiktok_music(url)
                    if not audio_path:
                        df.at[index, 'transcription'] = "Error downloading audio"
                        continue

                    # Step 2: Transcribe the audio
                    transcription = transcribe_audio(audio_path)
                    print(f"Row {index} Transcription: {transcription}")  # Debugging

                    # Step 3: Update the 'transcription' field
                    if transcription:
                        df.at[index, 'transcription'] = transcription
                    else:
                        df.at[index, 'transcription'] = "Empty transcription"

                except Exception as e:
                    print(f"Error processing row {index}: {e}")
                    df.at[index, 'transcription'] = "Error generating transcription"
            else:
                # Not a music URL
                print(f"Row {index} is not a TikTok music URL.")
                df.at[index, 'transcription'] = -1

    # Ensure the 'transcription' column has consistent data types
    df['transcription'] = df['transcription'].astype(str).fillna("")

    # Save the modified DataFrame back to Parquet
    df.to_parquet(output_parquet)

def process_all_dms_with_music_transcription(target_directory):
    for root, dirs, files in os.walk(target_directory):
        for file in files:
            if file == 'DMs.parquet':
                input_parquet_path = os.path.join(root, file)
                output_parquet_path = os.path.join(root, 'DMs.parquet')
                
                print(f"Processing {input_parquet_path}...")
                process_parquet_with_music_transcription(input_parquet_path, output_parquet_path)
                print(f"Saved processed file to {output_parquet_path}")
```
